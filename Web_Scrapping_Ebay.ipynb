```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://www.ebay.com/b/Cell-Phones-Smartphones/9355/bn_320094"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib')
table = soup.find('ul', class_='b-list__items_nofooter srp-results srp-grid') 
price = []
product = []
for i in table.findAll('li',class_='s-item s-item--large s-item--bgcolored'):
    price.append(i.find('div', class_='s-item__detail s-item__detail--primary').text)
    product.append(i.find('h3', class_='s-item__title').text)
    
df = pd.DataFrame(data={'Product':product,'Price':price})
```
